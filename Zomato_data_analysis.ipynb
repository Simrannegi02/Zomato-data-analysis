```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("/content/Zomato-data- (1).csv")
print(df)
```

```python
def clean_data(value):
  value=str(value).split('/')
  value=value[0];
  return float(value)
df['rate']  =df['rate'].apply(clean_data)
print(df.head())

```

```python
z=df.groupby("listed_in(type)")["votes"].sum()
result=pd.DataFrame({"votes":z})
plt.plot(result, c='green', marker='*')
plt.xlabel('types of resturant')
plt.ylabel('votes')
```
